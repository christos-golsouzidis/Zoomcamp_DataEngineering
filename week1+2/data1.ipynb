Question 5. Three biggest pickups

```python
zonedf = pd.read_csv('./taxi+_zone_lookup.csv')
```

```python
zonedf.head(10)
```

```python
zonedf.head(0).to_sql(name='zone_taxi_data',con = engine, if_exists='replace')
```

```python
zonedf.to_sql(name='zone_taxi_data',con = engine, if_exists='append')
```

```python
query = '''

select zone_taxi_data."Borough", sum(green_taxi_data."total_amount")
from zone_taxi_data
join green_taxi_data
on green_taxi_data."PULocationID" = zone_taxi_data."LocationID"
group by zone_taxi_data."Borough"
order by sum desc
'''
pd.read_sql(query, con=engine)
```

Question 6. Largest tip

```python
query = '''

select zone_taxi_data."Zone" , zone_taxi_data."LocationID" 
from zone_taxi_data
where "Zone"= 'Astoria'
'''

pd.read_sql(query, con=engine)
```

```python
query = '''

select green_taxi_data."tip_amount" , zone_taxi_data."Zone"
from green_taxi_data
join zone_taxi_data 
on green_taxi_data."DOLocationID" = zone_taxi_data."LocationID"
where green_taxi_data."PULocationID" = 7
order by "tip_amount" desc
'''
pd.read_sql(query, con=engine)
```
